## Matplotlib
A data visualizing library

#### Types of plot
1) Bar ===========> Makes a bar plot <br>
2) Barh ==========> Makes horizontal bar plot <br>
3) Boxplot =======> Make box plot and whisker plot <br>
4) hist ==========> Plot a histogram <br>
5) hist2d ========> Makes 2d histogram plot <br>
6) pie ===========> Plot a pie chart <br>
7) plot ==========> Plot lines <br>
8) Scatter =======> Makes Scatter plot <br>
9) Stackplot======> Draws a stacked area plot

#### Image Functions
1) imread =========> Read an image from a file into an array <br>
2) imsave =========> Save an array as in image file <br>
3) imshow =========> Display an image on axes <br>


#### Axes functions
1) axes ==========> Add axes to the figure <br>
2) text ==========> Add text to the axes <br>
3) title =========> Set the title of the current axes <br>
4) Xlabel ========> Set the axis label of the curent axis <br>
5) Xlim ==========> Get or set the x limits of the current axes <br>
6) Xticks ========> Get or set the x-limits of the current tick locations and labels <br>
Same for y-axis

### Box Plots

```python
np.random.seed(10)
collectn_1 = np.random.normal(100, 10, 200)
print(collectn_1)

fig = plt.figure()
# Create an axes instance
ax = fig.add_axes([0,0,1,1])
# Create the boxplot
bp = ax.boxplot(collectn_1)
plt.show()
```

The above figure explains that <br>
The Minimum value is around 78 <br>
the Maximum value  is around 125 <br><br>

The median value is 100 (Quartile) (50%) means 50% of the random variables are less than or equal to 100 <br>
The quartile (Q1) is around 95 means=====> 25% of the random variable are less than or equals to 95 <br>
The quartile (Q3) is around 105 means=====> 75% of the random variable are less than or equals to 105 <br>

### Generating random data and plotting in fig realtime

```python
import csv
import random
import time
import pandas as pd
```

```python
x_value = 0
total_1 = 1000
total_2 = 1000

field_names = ["x_value", "total_1", "total_2"]

with open('data.csv', 'w') as csv_file:
    csv_writer = csv.DictWriter(csv_file, fieldnames=field_names )
    csv_writer.writeheader()
    
while True:
    with open('data.csv', 'a') as csv_file:
        csv_writer = csv.DictWriter(csv_file, fieldnames=field_names)
        
        info = {
            "x_value": x_value,
            "total_1": total_1,
            "total_2": total_2
        }
        
        csv_writer.writerow(info)
        
        print(x_value,total_1,total_2)
        
        x_value += 1
        total_1 = total_1 + random.randint(-6,8)
        total_2 = total_2 + random.randint(-5,6)
    
    time.sleep(1)
```

### Multiplots

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) ## main axes
axes2 = fig.add_axes([0.55,0.55,0.3,0.3]) ## inset axes

y = np.sin(x)

axes1.plot(x,y,'y')
axes2.plot(x,np.cos(x),'r')

axes1.set_title('sine')
axes1.set_xlabel('medium')
axes1.set_ylabel('sales')

axes2.set_title('cosine')
axes2.set_xlabel('medium')
axes2.set_ylabel('sales')

plt.show()
```

### Matplotlib Sibplot function

```python
fig,a = plt.subplots(nrows=2,ncols=2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('Square')

a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('Square root')

a[1][0].plot(x,np.exp(x))
a[1][0].set_title('Exp')

a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')

plt.show()
```
