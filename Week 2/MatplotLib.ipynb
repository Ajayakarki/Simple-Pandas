## Matplotlib
A data visualizing library

#### Types of plot
1) Bar ===========> Makes a bar plot <br>
2) Barh ==========> Makes horizontal bar plot <br>
3) Boxplot =======> Make box plot and whisker plot <br>
4) hist ==========> Plot a histogram <br>
5) hist2d ========> Makes 2d histogram plot <br>
6) pie ===========> Plot a pie chart <br>
7) plot ==========> Plot lines <br>
8) Scatter =======> Makes Scatter plot <br>
9) Stackplot======> Draws a stacked area plot

#### Image Functions
1) imread =========> Read an image from a file into an array <br>
2) imsave =========> Save an array as in image file <br>
3) imshow =========> Display an image on axes <br>


#### Axes functions
1) axes ==========> Add axes to the figure <br>
2) text ==========> Add text to the axes <br>
3) title =========> Set the title of the current axes <br>
4) Xlabel ========> Set the axis label of the curent axis <br>
5) Xlim ==========> Get or set the x limits of the current axes <br>
6) Xticks ========> Get or set the x-limits of the current tick locations and labels <br>
Same for y-axis





```python
from matplotlib import pyplot as plt
import numpy as np
import math
```

```python
x = np.arange(1,math.pi*2,0.5)
y = np.sin(x)
print(x)
print(y)

plt.plot(x,y)
plt.xlabel('angle')
plt.ylabel('sine')
plt.title('Sin Wave')
plt.show()
```

4) 'D' ==========> Diamond marker <br>
5) 'H' ==========> Hexagon marker <br>

```python
y = [1, 4, 9, 16, 25,36,49, 64]
x1 = [1, 16, 30, 42,55, 68, 77,88]
x2 = [1,6,12,18,28, 40, 52, 65]
(underscore)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
l1 = ax.plot(x1,y,'ys--') ## solid line with yellow color and square marker
l2 = ax.plot(x2,y,'go--') ## dash line with reen colour ad circle marker

ax.legend(labels=('tv','smartphone'), loc = 'upper right') ## Legend place in lower right/upper right
ax.set_title("AAAAAAAAAAAAA")
ax.set_xlabel('medium')
ax.set_ylabel('sales')
plt.show()
```

### Multiplots

```python
x = np.arange(0, math.pi*2, 0.05)
fig = plt.figure()

axes1 = fig.add_axes([0.1,0.1,0.8,0.8]) ## main axes
axes2 = fig.add_axes([0.55,0.55,0.3,0.3]) ## inset axes

y = np.sin(x)

axes1.plot(x,y,'b')
axes2.plot(x,np.cos(x),'r')

axes1.set_title('sine')
axes1.set_xlabel('medium')
axes1.set_ylabel('sales')

axes2.set_title('cosine')
axes2.set_xlabel('medium')
axes2.set_ylabel('sales')

plt.show()
```

### Matplotlib Sibplot function

```python
fig,a = plt.subplots(2,2)
x = np.arange(1,5)
a[0][0].plot(x,x*x)
a[0][0].set_title('Square')
a[0][1].plot(x,np.sqrt(x))
a[0][1].set_title('Square root')
a[1][0].plot(x,np.exp(x))
a[1][0].set_title('Exp')
a[1][1].plot(x,np.log10(x))
a[1][1].set_title('log')
plt.show()
```

### Subplot2grid() Functions

```python
a1 = plt.subplot2grid((3,3),(0,0),colspan = 2)
a2 = plt.subplot2grid((3,3),(0,2), rowspan = 3)
a3 = plt.subplot2grid((3,3),(1,0),rowspan = 2, colspan = 2)

x = np.arange(1,10)
a1.plot(x, x*x)
a1.set_title('square')

a2.plot(x, np.exp(x))
a2.set_title('exp')

a3.plot(x, np.log(x))
a3.set_title('log')

plt.tight_layout()
plt.show()
```
