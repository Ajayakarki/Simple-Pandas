### What % of people from each country know Python?

### 1st finding Total number of respondents of each country

```python
country_respondents = df['Country'].value_counts()
country_respondents
```

### 2nd Finding total number of people who knows python

```python
country_uses_python = country_group['LanguageWorkedWith'].apply(lambda x: x.str.contains('Python').sum())
country_uses_python
```

### Now concanating two series of dataframe

```python
python_dataframe = pd.concat([country_respondents, country_uses_python], axis=1, sort=False)
python_dataframe
```

### Now Renaming columns

```python
python_dataframe.rename(columns={'Country': 'TotalRespondents', 'LanguageWorkedWith': 'KnowsPython'}, inplace=True)
```

### Last Finding the % of people from each country who knoes python

```python
python_dataframe.head()
```

```python
python_dataframe['Total %'] = (python_dataframe['KnowsPython'] / python_dataframe['TotalRespondents'] * 100 )
```

```python
python_dataframe
```

```python
python_dataframe.sort_values(by='Total %', ascending=False).head(20)
```
